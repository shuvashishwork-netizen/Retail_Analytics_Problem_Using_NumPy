**Retail Analytics Problem Using NumPy8**



A retail chain tracks weekly sales for 4 products across 3 stores. The company also has cost prices, selling prices, and wants to compute profit, performance, and store rankings.

```python
import numpy as np
```

**Given Data**

1. Weekly units sold (rows = stores, columns = products)

```python
sales_unit = np.array([
    [120, 80, 60, 40], #sales for Store 1
    [100, 90, 50, 30], #sales for Store 2
    [140, 70, 65, 55]  #sales for Store 3
])
```

2. Selling price per product (flat vector)


```python
selling_price = np.array(
    [20, 15, 10, 8]
)
```

3. Cost price per product (flat vector)


```python
cost_price = np.array([12, 9, 6, 4])
```
